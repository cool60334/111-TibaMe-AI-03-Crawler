## 爬取PChome搜尋結果

````python
import requests
import prettytable
import IPython.display as dp

def search(keyword=None, page="1"):

  # 輸入要搜尋的關鍵字

  keyword = str(input("關鍵字: "))
  

  while True:

    # 建立表格，設定靠左對齊

    table = prettytable.PrettyTable(["名稱", "價格"], encoding = "utf-8-sig")
    table.align["名稱"] = "l"
    table.align["價格"] = "l"


    # 取得pchome的JSON檔

    req = requests.get("https://ecshweb.pchome.com.tw/search/v3.3/all/results",
                      params = {
                          "q": keyword,
                          "page": page,
                          "sort": "rnk/dc"
                      })

    res = req.json()


    # 將資料放到表格中並印出

    for r in res["prods"]:
      table.add_row([r["name"], r["price"]])

    print(table)


    # 輸入要前往的頁碼

    page = str(input("前往頁碼: "))


    # 去除先前的輸出結果

    dp.clear_output(True)
````

## 抓取勞動部網站資料

````python
import requests
from bs4 import BeautifulSoup
````

````python
url = "https://www.mol.gov.tw/1607/1632/1633/lpsimplelist"
req = requests.get(url)

# 取得網頁html的文字

html = req.text

# 將html文字解析成html格式

soup = BeautifulSoup(html, "html.parser")

# 找到所有<h3>標籤，結果為一個list

title_list = soup.find_all("h3")

# 方法一: 使用.text找出<h3>標籤內包覆的文字

# for t in title_list:
#   print(t.text)


# 方法二: 使用<h3>標籤內的<a>標籤的屬性抓取文字
# .attrs -> 表示找到標籤中全部的屬性

for t in title_list:
  title = t.find("a").attrs["title"]
  print(title)
````

````python
# 抓取標題、連結、發布日期、點閱次數

url = "https://www.mol.gov.tw/1607/1632/1633/lpsimplelist"
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, "html.parser")

content_list = soup.find_all("div", {"class": "item_list2"})

for content in content_list:
  title = content.find("a").attrs["title"]  # 標題
  link = content.find("a").attrs["href"]  # 連結

  info = content.find("div", {"class": "data"})
  pub_date = info.find_all("span")[2].text  # 發布日期
  click = info.find_all("span")[3].text  # 點閱次數

  print(f"標題: {title}\n連結: https://www.mol.gov.tw{link}\n{pub_date}\n{click}\n")
````

## 抓取外交部網站新聞

### 使用.parent

````python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://www.mofa.gov.tw/News.aspx",
                   {
                    "n": "96",
                    "sms": "74",
                    "page": "1",
                    "PageSize": "20"   
                   })

soup = BeautifulSoup(req.text, "html.parser")

# 先找到含有日期與標題的程式碼的list

news = soup.find_all("td", {"class": "CCMS_jGridView_td_Class_0 is-center"})


for n in news:
  
  # 取得日期標籤的文字

  date = n.find("span").text

  """
  使用n.parent先找到包著("td", {"class": "CCMS_jGridView_td_Class_0 is-center"})的標籤<tr>
  再從<tr>標籤取得包在<td>標籤內的標題文字
  """

  title = n.parent.find("td", {"class": "CCMS_jGridView_td_Class_1 is-left"}).text
  link = "https://www.mofa.gov.tw/" + n.parent.find("a").attrs["href"]

  # 取得新聞內容

  content_req = requests.get(link)
  c_soup = BeautifulSoup(content_req.text, "html.parser")

  article = c_soup.find("div", {"class": "p"}).find("span").text

  print(date, "\n", title, "\n", link, "\n\n", article, sep = "")

  # 只抓第一個
  
  break
````

# 爬取Tabelog網頁

使用urllib爬取網頁，可搭配HTTPError，處理爬取過程中出現的網頁錯誤

````python
from urllib.request import urlopen
from urllib.error import HTTPError  # 處理網頁的錯誤訊息
from bs4 import BeautifulSoup
import pandas as pd


df = pd.DataFrame(columns = ["en_name", "jp_name", "rate", "link"])

page = 51

while True:
  url = "https://tabelog.com/tw/tokyo/rstLst/" + str(page) + "/?SrtT=rt"

  print(f"第 {page} 頁: {url}")

  try:
    req = urlopen(url)

  except HTTPError:
    print("已經最後一頁了")
    break
  
  soup = BeautifulSoup(req, "html.parser")
  res = soup.find_all("li", {"class": "list-rst js-list-item"})

  for r in res:
    en_name = r.find("a", {"class": "list-rst__name-main js-detail-anchor"}).text
    jp_name = r.find("small", {"class": "list-rst__name-ja"}).text
    rate = r.find("b", {"class": "c-rating__val"}).text
    link = r.find("a", {"class": "list-rst__name-main js-detail-anchor"}).attrs["href"]  # 也可以直接r.find(...)["href"]
    
    print(en_name, jp_name, rate, link)

    s = pd.Series([en_name, jp_name, rate, link], index = ["en_name", "jp_name", "rate", "link"])
    df = df.append(s, ignore_index = True)

  page = page + 1

df.to_csv("tabelog.csv", encoding = "utf-8-sig", index = False)
````

## 使用requests (較簡潔)

````python
import requests
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/movie/M.1676618759.A.6EC.html"
req = requests.get(url)

soup = BeautifulSoup(req.text, "html.parser")
content = soup.find("div", {"id": "main-content"})

# 取得作者、看板、標題、時間

metas = content.find_all("span", {"class": "article-meta-value"})
print("作者:", metas[0].text)
print("看板:", metas[1].text)
print("標題:", metas[2].text)
print("時間:", metas[3].text)


# 刪掉不需要的元素(作者、標題、時間)

metalines = content.find_all("div", {"class", "article-metaline"})
for m in metalines:
  m.extract()

# 刪掉不需要的元素(看板)

metalines = content.find_all("div", {"class", "article-metaline-right"})
for m in metalines:
  m.extract()

# 計算分數

score = 0

pushes = content.find_all("div", {"class": "push"})

for p in pushes:
  tag = p.find("span", {"class": "push-tag"}).text

  if "推" in tag:
    score += 1
  elif "噓" in tag:
    score -= 1

  p.extract()

print("分數：", score)
print("本文：", content.text)
````
